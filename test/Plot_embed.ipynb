```python
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt
```

```python
embed_vec = np.load('embed.npy')

embed_vec
```

```python
model = TSNE(n_components=3)
```

```python
transformed_vec = model.fit_transform(embed_vec)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs=transformed_vec[:, 0], ys=transformed_vec[:, 1], zs=transformed_vec[:, 2])

plt.show()
```
